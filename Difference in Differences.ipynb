# Case 2

Ths case show wheatehr tax credit has ositive impact on simgle mum employment

```python
df2=pd.read_stata("eitc.dta")
```

```python
df2
```

```python
#make dummy variable for before 1993 and post 1993
df2['post93']=np.where(df2['year']>1993, 1, 0)
```

```python
df2
```

```python
#make dummy variable for single mom or not
df2['mom']=np.where(df2['children']>0, 1, 0)
```

```python
df2
```

```python
# create intersection
df2['mom_post93']=df2['post93']*df2['mom']
```

```python
df2
```

```python
Y = df2.loc[:,'work'].values
X = df2.loc[:,['post93','mom','mom_post93']].values
```

```python
X
```

```python
#logistic regression
import statsmodels.api as sm
X=sm.add_constant(X)
model2=sm.Logit(Y,X).fit()
model2.summary(yname="Work",xname=("intercept","After 1993","Is Mom", "Mom after 1993"))
```

-  employment has no change after 1993 as p value is 0.87 which is over 0.05
-  Mom emplotment has decreased overall period as pvalue is less than 1%
-  mom after 1993 employment has increased as p value is less than 0.05
-  therefore tax credit has positive effect on single mom employment

```python
Y = df2.loc[:,'work'].values
X = df2.loc[:,['post93','mom','mom_post93','nonwhite','ed']].values
```

```python
X=sm.add_constant(X)
model3=sm.Logit(Y,X).fit()
model3.summary(yname="Work",xname=("intercept","After 1993","Is Mom", "Mom after 1993",'nonwhite','year of education'))
```

- employment has no change after 1993 as p value is 0.87 which is over 0.05
- Mom emplotment has decreased overall period as pvalue is less than 1%
- still mom after 1993 employment has increased as p value is 0.05, which is statistically significant
- therefore tax credit has positive effect on single mom employment
